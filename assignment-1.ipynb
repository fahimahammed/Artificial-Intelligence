Assignment 1: Write a Python program to display values of y when 'x' is an independent variable having values [-10, 10, step_size = 0.1]

- y = wx + b; where 'w' and 'b' are constants
- y = x^2
- y = 1 / (1 + e^(-x))
- y = (e^x - e^(-x)) / (e^x - e^(-x))
- y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))
- y = g(f(x)) where u = f(x) = wx + b and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))
- y = g3(f3(v)) where w = f3(v) = w3y1 + w4y2 + b & g3(w) = 1 / (1 + e^(-w)) | y1 = g1(f1(x)) where u1 = f1(x) = w1x + b1 & g1(u1) = 1 / (1 + e^(-u1)) | y2 = g2(f2(x)) where u2 = f2(x) = w2x + b2 & g2(u2) = 1/ (1 + e^(-u2))

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function1(x, w, b):
    return w*x+b

def function2(x):
    return x**2

def function3(x):
    return 1 / (1+np.exp(-x))

def function4(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) - np.exp(-x))

def function5(x, w, b):
    return function3(-function1(x, w, b))

def function6(x, w, b):
    return (np.exp(function1(x, w, b)) - np.exp(-function1(x,w,b))) / (np.exp(function1(x,w,b)) + np.exp(-function1(x,w,b)))

def function7(x, w, b):
    return w*x+b
```

```python
def plot(rows, cols, num, x_val, y_val, title):
    plt.subplot(rows, cols, num)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(title)
    plt.plot(x_val, y_val)
    plt.tight_layout()
```

```python
x = np.arange(-10, 10.1, 0.1)
# print(x)
len(x)

w = 2
b = 3
plt.figure(figsize=(12, 16))
y_val = function1(x, w, b)
plot(4, 2, 1,x, y_val, "y = wx + b")

y_val = function2(x)
plot(4, 2, 2, x, y_val, "y = x^2")

y_val = function3(x)
plot(4, 2, 3, x, y_val, "y = 1 / (1 + e^(-x))")

y_val = function4(x)
plot(4, 2, 4, x, y_val, "y = (e^x - e^(-x)) / (e^x + e^(-x))")

y_val = function5(x,w,b)
plot(4,2,5, x, y_val, "y = g(f(x))")

y_val = function6(x, w, b)
plot(4,2,6, x, y_val, "y = g(f1(x)) + g(f2(x))")

y_val = function7(x,w,b)
plot(4,2,7, x, y_val, "y = g(f1(x)) + g(f2(x))")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Variables
x = np.arange(-10, 10, 0.1, dtype = np.float64)
w = 7
b = 4
figure_size = (12, 8)
```

```python
#Function: y = wx + b
y = w * x + b

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = wx + b")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "o", label = "y = f(x) = wx + b")
plt.grid()
plt.legend()
plt.show()
```

```python
# Function: y = x^2
y = pow(x, 2)

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "o", label = "y = f(x) = y = x^2")
plt.grid()
plt.legend()
plt.show()
```

```python
#Function: y = 1 / (1 + e^(-x))
y = 1 / (1 + np.exp(-x))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = 1 / (1 + e^(-x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "o", label = "y = f(x) = y = 1 / (1 + e^(-x))")
plt.grid()
plt.legend()
plt.show()
```

```python
# Function: y = (e^x - e^(-x)) / (e^x - e^(-x))
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) - np.exp(-x))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = y = (e^x - e^(-x)) / (e^x - e^(-x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "o", label = "y = f(x) = y = y = (e^x - e^(-x)) / (e^x - e^(-x))")
plt.grid()
plt.legend()
plt.show()
```

```python
#Function: y = g(f(x)) where u = f(x) = wx + b & y = g(u) = 1/ (1 + e^(-u))
u = w * x + b
y = 1 / (1 + np.exp(-u))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = g(f(x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "o", label = "y = f(x) = g(f(x))")
plt.grid()
plt.legend()
plt.show()
```

```python
#Function: y = g(f(x)) where u = f(x) = wx + b & g(u) = (e^u - e^(-u)) / (e^u + e^(-u))
u = w * x + b
y = (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = g(f(x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "o", label = "y = g(f(x))")
plt.grid()
plt.legend()
plt.show()
```

```python
#Function: y = g3(f3(v)) where w = f3(v) = w3y1 + w4y2 + b & g3(w) = 1 / (1 + e^(-w)) | y1 = g1(f1(x)) where u1 = f1(x) = w1x + b1 & g1(u1) = 1 / (1 + e^(-u1)) | y2 = g2(f2(x)) where u2 = f2(x) = w2x + b2 & g2(u2) = 1/ (1 + e^(-u2))
w1 = -3
b1 = -5
u1 = w1 * x + b1
y1 = 1 / (1 + np.exp(-u1))

w2 = 3
b2 = 5
u2 = w2 * x + b2
y2 = 1 / (1 + np.exp(-u2))

w3 = 4
w4 = 7
y = 1 / (1 + np.exp(-w3 * y1 - w4 * y2 - b))

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = g3(f3(x))")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "o", label = "y = g3(f3(x))")
plt.grid()
plt.legend()
plt.show()
```

```python
# Function y = wsin(x) + b
y = w * np.sin(x) + b

plt.figure(figsize = figure_size, dpi = 100)
plt.title("Function : y = wsin(x) + b")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, marker = "o", label = "y = f(x) = wsin(x) + b")
plt.grid()
plt.legend()
plt.show()
```
